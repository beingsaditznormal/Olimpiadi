```python
import mysql.connector

mydb = mysql.connector.connect(
  host="localhost",
  user="root",
  password="1234",
  database="ol"
)
```

```python
mycursor = mydb.cursor()

mycursor.execute("select nome, cognome, nazione from atleta")

myresult = mycursor.fetchall()

for x in myresult:
  print(x)
```

```python
mycursor = mydb.cursor()

mycursor.execute("SELECT nazione, COUNT(*) AS numero_atleti FROM atleta GROUP BY nazione ORDER BY numero_atleti DESC;")

myresult = mycursor.fetchall()

for x in myresult:
  print(x)
```

```python
mycursor = mydb.cursor()

mycursor.execute("SELECT evento.id_evento, evento.codice, sport.nome_sport  FROM evento  JOIN sport ON evento.id_sport = sport.id_sport")

myresult = mycursor.fetchall()

for x in myresult:
  print(x)
```
